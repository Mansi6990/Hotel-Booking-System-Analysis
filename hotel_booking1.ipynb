##Data cleaning

```python
#checking dataset for duplicate value count
df_copy.duplicated().value_counts()
#True means duplicate rows
```

```python
#visualising duplicate values through count plot
plt.figure(figsize = (10,5))
sns.countplot(x=df_copy.duplicated())
```

```python
#dropping duplicate values
df_copy.drop_duplicates(inplace=True)
```

```python
df_copy.shape #updated rows and columns after dropping the duplicate values
```

```python
#checking for null/missing values
df_copy.isnull().sum().sort_values(ascending=False)
```

```python
#visualising the null/missing values through heatmap
plt.figure(figsize=(25,10))
sns.heatmap(df_copy.isnull(), cbar=False, yticklabels=False, cmap='viridis')
plt.title("Places of missing values in column")
```

So we have 4 columns with null values in them - company, agent, country and children
1. For children, nulls values will be replaced by 0 assuming no children.
2. For country columns. the null values will be replaced by 'others' assuming the country was not mentioned by the customer while booking.
3. For both company and agent column, the null values will be replaced by 0, as the customers might not have booked through them.



```python
#Replacing null/missing values with 0
null_columns= ['company','agent','children']
for col in null_columns:
  df_copy[col].fillna(0,inplace=True)

#Replacing NA values with 'others'
df_copy['country'].fillna('others',inplace=True)
```

```python
df_copy.isnull().sum().sort_values(ascending=False)
```

Here we can see the null values are handled successfully by replacing them with 0/'others'

```python
#Number of rows where the sum of adults, children and babies is 0, that means no bookings were made
len(df_copy[df_copy['adults']+df_copy['children']+df_copy['babies']==0])
```

```python
#Dropping all 166 rows
df_copy.drop(df_copy[df_copy['adults']+df_copy['children']+df_copy['babies']==0].index,inplace=True)
```

```python
#checking the unique values in columns
#creating a series consisting every column name of the dataset and it's value
#by using for loop to iterate over every column in the dataset
pd.Series({col:df_copy[col].unique() for col in df_copy})
```

```python
#Adding some new columns
df_copy['total_guests'] = df_copy['adults'] + df_copy['children'] + df_copy['babies']
df_copy['total_stay'] = df_copy['stays_in_week_nights'] + df_copy['stays_in_week_nights']
df_copy.shape
```

# 7. Which distribution channel is most used in booking?

```python
#creating a dataset of distribution channel name and count and renaming columns for better understanding
dist_df = df_copy['distribution_channel'].value_counts().reset_index()
dist_df=dist_df.rename(columns={'index': 'channel name', 'distribution_channel':'channel count'})
```

```python
#creating a explode data
my_explode=(0.05,0.05,0.05,0.05,0.05)

#adding percentage columns to the distribution channel dataset
dist_df['percentage']=round(dist_df['channel count']*100/df_copy.shape[0],1)

#setting the figure size
plt.figure(figsize=(25,10))

#plotting the chart values
plt.pie(dist_df['channel count'],labels=None, explode=my_explode,startangle=50)

#adding legends with percentage using list comprehension
labels=[f'{l},{s}%' for l,s in zip(dist_df['channel name'].value_counts().index.tolist(), dist_df['percentage'].values.tolist())]
plt.legend(bbox_to_anchor=(0.85,1),loc='upper left',labels=labels)

#setting the title
plt.title('Most used distribution channels for booking',fontsize=20)

#show the chart
plt.axis('equal')
plt.show()
```

###**Observation**
From this pie-chart visualisation we can see that TA/TO(Travel agents/tour operators) is the most used distribution channel by guests.

#19. What is the adr different months?

```python
#grouping arrival_month and hotel on mean of adr
booking_month=df_copy.groupby(['arrival_date_month','hotel'])['adr'].mean().reset_index()

#creating a month list to order the months in ascending
months=['January','February','March','April','May','June','July','August','September','October','November','December']

#creating a dataset of months, hotel and their adr
booking_month['arrival_date_month']=pd.Categorical(booking_month['arrival_date_month'],categories=months, ordered=True)

#sorting the months
booking_month=booking_month.sort_values('arrival_date_month')
booking_month
```

```python
#setting the chart
plt.figure(figsize=(15,5))

#plotting the values in a line chart
sns.lineplot(x=booking_month['arrival_date_month'],y = booking_month['adr'], hue=booking_month['hotel'])

#setting the labels and title
plt.title('ADR across each month', fontsize=20)
plt.xlabel('Month Name',fontsize=10)
plt.ylabel('ADR',fontsize=10)

#show
plt.show()
```

###**Observation**
**City hotel:** We can see that city hotel generates more revenue in May than other months.     
**Resort hotel:** In resort hotel the maximum revenue is generated between July and August.
